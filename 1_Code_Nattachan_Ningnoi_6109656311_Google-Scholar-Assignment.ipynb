### Author Table

#### เรียกใช้ WebDriver

```python
# เรียกใช้ WebDriver ซึ่งอยู่ใน current directory
driver = webdriver.Chrome("./chromedriver")
# เรียกใช้เว็บไซต์ที่เราต้องการดึง
driver.get("https://scholar.google.com/citations?view_op=view_org&org=10241031385301082500&hl=en&oi=io")
```

```python
%%time
'''
%%time คือ magic command ที่ใช้ในการติดตามสิ่งที่เรากำลังประมวลผลอยู่ทั้งหมดภายใน Cell นั้น ๆ โดยจะบอกถึง [100]
 CPU times: 
  • user คือ เวลาทั้งหมดที่ใช้โดย CPUs ในระหว่างการคำนวณ
  • sys  คือ เวลาทั้งหมดที่ใช้โดย CPU ในระหว่างการทำงานที่เกี่ยวข้องกับระบบ เช่น การจัดสรรหน่วยความจำ (memory allocation)
 Wall time: 
  • คือเวลาทั้งหมดที่นับตั้งแต่เริ่มรันโปรแกรมจนถึงสิ้นสุดโปรแกรมหรือความแตกต่างระหว่างเวลาที่งานเสร็จสิ้นและเวลาที่งานเริ่มต้นขึ้น
'''
# สร้าง data frame ที่ใช้ในการเก็บข้อมูลของ Author
author_df = pd.DataFrame(columns=['user_ID','name','affiliation'])
'''
 • 'user_ID' ใช้ในการเก็บรหัสของ Author หรือผู้ใช้งานนั้น ๆ
 • 'name' ใช้ในการเก็บชื่อของ Author
 • 'affiliation' หรือ สังกัด ใช้ในการเก็บข้อมูลเกี่ยวกับคณะหรือสังกัดของ Author
'''

# กำหนด row = 1
'''
กำหนด row = 1 เพื่อกำหนดแถวของ data frame ที่เราจะเก็บข้อมูลลงไป โดยเริ่มแถวต้นคือคือ 1 แล้วเพิ่มขึ้นเรื่อย ๆ 
เมื่อเก็บข้อมูลครบ 1 ครั้ง จำนวนแถวก็จะเพิ่มขึ้น จาก row +=1
'''
row = 1

while True: 
    '''
    โดยเราจะใช้ While loop ซึ่งเป็นคำสั่งวนซ้ำ คือ เราจะเก็บข้อมูลทั้งหมด 
    จนกว่าเงื่อนไขจะเป็นจริง คือ เก็บข้อมูลครบทั้งหมด 
    '''

    author_objects = driver.find_elements_by_class_name('gsc_1usr')
    '''กำหนดตัวแปร author_objects ซึ่งเปรียบเสมือนกล่อง ๆ หนึ่ง 
    แล้วในกล่องก็จะมีสิ่งของต่าง ๆ ได้แก่ user_ID , author_name และ affiliation
    ซึ่งอยู่ใน class ที่มีชื่อว่า 'gsc_1usr'
    '''    
    
    # ใช้ for loop ในการเก็บค่าต่าง ๆ ที่อยู่ในกล่องใบนั้น ได้แก่ user_ID , author_name และ affiliation
    for item in author_objects:
        link    = item.find_element_by_class_name('gs_ai_pho')
        '''
        Link โดยเราจะใช้ find_element_by_class_name('class_name) เพื่อใช้ในการหา class name 
        ที่มีชื่อว่า 'gs_ai_pho' ถ้ามี class name ที่ตรงกันเราจะได้ค่าที่อยู่ใน class นั้นออกมาคือ Link ของ 
        Author แต่ละคน ตัวอย่างเช่น href="/citations?hl=en&user=97Zz_TIAAAAJ" 
        '''
        user_ID = str(link.get_attribute('href')).replace('https://scholar.google.com/citations?hl=en&user=','')
        author_name = item.find_element_by_class_name('gs_ai_name')
        author_name = author_name.text
        '''
        author_name โดยเราจะใช้ find_element_by_class_name('class_name) เพื่อใช้ในการหา class name ที่มีชื่อว่า 'gs_ai_name' 
        ถ้ามี class name ที่ตรงกันเราจะได้ค่าที่อยู่ใน class นั้นออกมาคือ ชื่อของ Author แล้วแปลง element ต่าง ๆ ให้อยู่ในรูปแบบของ text
        '''
        
        affiliation = item.find_element_by_class_name('gs_ai_aff')
        affiliation = affiliation.text
        '''
        affiliation จะทำรูปแบบเดียวกันกับ author_name การหา class name ที่ตรงกันคือ 'gs_ai_aff' 
        แล้วค่าที่อยู่ใน class นั้นให้อยู่ในรูปแบบของ text
        '''
        
        # เพิ่มข้อมูลลงในไฟล์ 
        author_df.loc[row] =[user_ID,author_name,affiliation]
        # เพิ่มจำนวน row เพื่อบอก data frame ให้ขึ้น row ใหม่
        row +=1
    
    # หา URL ปัจจุบันที่ webdriver กำลังเรียกใช้งานอยู่
    page = requests.get(driver.current_url)
    # ส่งคืนค่าเนื้อหาในเว็บกลับมาในรูปแบบของ raw bytes
    soup = BeautifulSoup(page.content)
    
    # ใช้ในการคลิกปุ่มเพื่อเปลี่ยนหน้า
    next_buttons = soup.find_all(class_='gs_btnPR')
    if next_buttons[0].has_attr('onclick'):
        driver.find_element_by_css_selector(".gs_btnPR").click()
    else:
        break
    # หยุด โดยนับ 0 ถึง 3 แล้วค่อยเริ่มต้นทำงานใหม่ 
    sleep(randint(0,5))
    
# save file ในรูปเเบบ .CSV
author_df.to_csv('author.csv')
# เพื่อปิดหน้าต่างเว็บ browser ที่ webdriver กำลังเรียกใช้งานอยู่
driver.close()
```

## Paper Table

```python
%%time

# เรียกใช้ webdriver.Chrome()
driver = webdriver.Chrome("./chromedriver")
# เรียกใช้ไฟล์ author.csv เพื่อดึงข้อมูล 'user_ID'
author = pd.read_csv("author.csv")
# สร้าง DataFrame ที่ใช้เก็บ Paper
paper_df = pd.DataFrame(columns=['title','authors','publication_date','description','cite_by'])

# กำหนด row = 1 เพื่อใช้ในการเพื่อกำหนดแถวของ data frame 
row = 1
# ใช้ในการลำดับของ author ที่เรากำลังดึงอยู่
user_queue = 0

# for loop วนซ้ำตามจำนวนของ user_ID
for user_ID in author['user_ID']:
    # ใช้ webdriver ในการเปิดเว็บ browser ตามด้วย user_ID ที่เราเก็บไว้ในไฟล์ author.csv" 
    driver.get('https://scholar.google.com/citations?hl=en&user='+ user_ID)
    # ค้นหา ปุ่ม SHOW MOER 
    button_show_more = driver.find_elements_by_class_name('gs_btnPD')
    # เมื่อเจอปุ่ม SHOW MOER  ให้คลิกปุ่มนั้นจนกว่าปุ่มนั้นจะกดไม่ได้ หรือ เเสดง paper ครบหมดเเล้ว
    while button_show_more[0].is_enabled()==True:
        # คลิกที่ปุ่ม SHOW MOER
        driver.find_element_by_class_name('gs_btnPD').click()
        # เมื่อคลิกแล้วให้รอนับ 0 ถึง 5
        sleep(randint(0,5))
    
    # Link ที่บอกรายละเอียดของ paper นั้นๆ
    link = driver.find_elements_by_class_name('gsc_a_at')
    # เลือกเฉพาะ link ออกมา
    link = [i.get_attribute('data-href') for i in link]
    
    # เเสดงคิวของ Author ที่เราทำการดึงข้อมูล
    print("------------------------------------------------")
    print("User Queue : ",user_queue)
    print(" ")
    user_queue += 1

    for url in link:
        driver.get('https://scholar.google.com/'+url)
        try :
            table = driver.find_element_by_id('gsc_vcd_table')
            body = table.find_elements_by_class_name('gs_scl')
            
            Title   = " "
            Authors = " "
            Publication_date = " "
            Description = " "
            cited = " "
            
            for item in body:
                Title = driver.find_elements_by_class_name('gsc_vcd_title_link')[0].text
                if item.find_elements_by_class_name('gsc_vcd_field')[0].text == 'Authors':
                    Authors = item.find_elements_by_class_name('gsc_vcd_value')[0].text
                elif item.find_elements_by_class_name('gsc_vcd_field')[0].text == 'Publication date':
                    Publication_date = item.find_elements_by_class_name('gsc_vcd_value')[0].text
                elif item.find_elements_by_class_name('gsc_vcd_field')[0].text == 'Description':
                    Description = item.find_elements_by_class_name('gsc_vcd_value')[0].text
                elif item.find_elements_by_class_name('gsc_vcd_field')[0].text == 'Total citations':
                    cited = item.find_elements_by_css_selector('a')[0].text.replace('Cited by ','')
            paper_df.loc[row]=[Title,Authors,Publication_date,Description,cited]
            row += 1
        except :
            pass
        print('Articles number >>> ',len(paper_df))
        sleep(randint(0,5))
paper_df.to_csv('paper.csv')
driver.close()  
```
